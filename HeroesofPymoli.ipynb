```python
import pandas as pd
```

```python
file_to_load = "/Users/hallemorel/Documents/Resources/purchase_data.csv"
```

```python
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
# getting ridding of duplicates so as to not have them show up in the data frame
players = purchase_data.loc[:, ["Gender", "SN", "Age"]]
players = players.drop_duplicates()
number_players = players.count()[0] 
pd.DataFrame({"Total Players": [number_players]})
```

```python
#using the purchase_data to calculate the average, total, count and unique values each of the following to form the table
avg_item_price = purchase_data["Price"].mean()
total_purchase = purchase_data["Price"].sum()
purchase_count = purchase_data["Price"].count()
item_count = len(purchase_data["Item ID"].unique())

table = pd.DataFrame ({"Number of Unique Items": [item_count],
                      "Total Revenue": [total_purchase],
                      "Number of Purchases": [purchase_count],
                      "Average Price": [avg_item_price]})
table
```

```python
#calculating the following values from the purchase_data info but by gender
gender_count_total = players["Gender"].value_counts()
gender_percent_total = gender_count_total/number_players
gender = pd.DataFrame({"Total Count": gender_count_total, "Percentage of Players": gender_percent_total})
gender
```

```python
#Creating a more specific table in order to gather more values and information based on the genders
gender_purchase = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
gender_avg = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_count = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
normalized_total = gender_purchase / gender["Total Count"]
gender_data = pd.DataFrame({"Purchase Count": gender_count, "Average Purchase Price": gender_avg, "Total Purchase Value": gender_purchase, "Normalized Total": normalized_total})
gender_data
```

```python
#using bins to break down the percentage of players and total count by age, helps creating groupings based on visually where the data lies, not incrementally
age_bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
players ["Age Ranges"] = pd.cut(players["Age"], age_bins, labels=group_names)
age_totals = players ["Age Ranges"].value_counts()
age_percents = age_totals/number_players
age = pd.DataFrame({"Total Count": age_totals, "Percentage of Players": age_percents})
age = age.sort_index()
age
```

```python
#continuatio of previous data but adding more columns and values
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

age_purchase_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
age_average = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
age_counts = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")

normalized_total = age_purchase_total / age ["Total Count"]

age_data = pd.DataFrame ({"Purchase Count": age_counts, "Average Purchase Price": age_average, "Total Purchase Value": age_purchase_total, "Normalized Totals": normalized_total})
age_data
```

```python
#now evaluating the data based on the SN column, calculating count, average... etc
user_purchase_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
user_avg_purchase_price = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_total_purchase_value = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

user_summary = pd.DataFrame ({"Purchase Count":user_purchase_count,"Average Purchase Price": user_avg_purchase_price,"Total Purchase Value":user_total_purchase_value})
sorted_descending = user_summary.sort_values("Total Purchase Value", ascending = False)

sorted_descending.head()
```

```python
item_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
item_data
```

```python
#this table is a breakdown by the item ID
item_purchase_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
item_price = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Item Price")
total_item_value = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

item_pd = pd.DataFrame({"Purchase Count":item_purchase_count, "Item Price":item_price,"Total Purchase Value":total_item_value})

item_sorted = item_pd.sort_values("Purchase Count", ascending=False)

item_sorted
```

```python
#shifts the data so that it is in descending order
item_total_purchase = item_pd.sort_values("Total Purchase Value", ascending=False)
item = item_total_purchase.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]
item
```
